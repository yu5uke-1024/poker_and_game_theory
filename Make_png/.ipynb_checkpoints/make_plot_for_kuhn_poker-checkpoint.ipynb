```python
import pandas as pd 
import matplotlib.pyplot as plt

# -- player2 --
df2_1 = pd.read_csv("output/database_for_plot_2_42.csv")
df2_2 = pd.read_csv("output/database_for_plot_2_1000.csv")
df2_3 = pd.read_csv("output/database_for_plot_2_10000.csv")
# -- player3 --
df3_1 = pd.read_csv("output/database_for_plot_3_42.csv")
df3_2 = pd.read_csv("output/database_for_plot_3_1000.csv")
df3_3 = pd.read_csv("output/database_for_plot_3_10000.csv")
# -- player4 --
df4_1 = pd.read_csv("output/database_for_plot_4_42.csv")
df4_2 = pd.read_csv("output/database_for_plot_4_1000.csv")
df4_3 = pd.read_csv("output/database_for_plot_4_10000.csv")

# -- player5 --
df5_1 = pd.read_csv("output/database_for_plot_5_42.csv")
df5_2 = pd.read_csv("output/database_for_plot_5_1000.csv")
df5_3 = pd.read_csv("output/database_for_plot_5_10000.csv")


df = pd.concat([df2_1, df2_2, df2_3, df3_1, df3_2, df3_3, df4_1, df4_2, df4_3, df5_1, df5_2, df5_3], axis=1, join='outer')

df = df.loc[:,~df.columns.duplicated()]
df = df.set_index('iteration')
df["exploitability_rate_2_mean"] = df[["exploitability_rate_2_42", "exploitability_rate_2_1000", "exploitability_rate_2_10000"]].mean(axis=1)
df["exploitability_rate_3_mean"] = df[["exploitability_rate_3_42", "exploitability_rate_3_1000", "exploitability_rate_3_10000"]].mean(axis=1)
df["exploitability_rate_4_mean"] = df[["exploitability_rate_4_42", "exploitability_rate_4_1000", "exploitability_rate_4_10000"]].mean(axis=1)
df["exploitability_rate_5_mean"] = df[["exploitability_rate_5_42", "exploitability_rate_5_1000", "exploitability_rate_5_10000"]].mean(axis=1)

df.tail()
```

```python
plt.rcParams["font.size"] = 40
plt.figure(figsize=(24,12))
plt.plot(df.index, df.exploitability_rate_2_mean, label="2-Kuhn exploit" , lw=5, color="r")
plt.plot(df.index, df.exploitability_rate_3_mean, label="3-Kuhn exploit", lw=5 color="dodgerblue")
plt.plot(df.index, df.exploitability_rate_4_mean, label="4-Kuhn exploit" , lw=5, color = "limegreen")
plt.plot(df.index, df.exploitability_rate_5_mean, label="5-Kuhn exploit", lw=5, color="magenta")

plt.xscale('log')
plt.yscale('log')
plt.xlabel('iteration')
plt.ylabel('exploitability rate')
plt.legend(loc = "lower left")
plt.grid()
plt.xlim(10)
plt.ylim(0.001,10)
plt.savefig("./png/kuhn_poker_nfsp.png")
```
