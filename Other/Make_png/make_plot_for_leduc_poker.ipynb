```python
import pandas as pd 
import matplotlib.pyplot as plt

df1 = pd.read_csv("output/database_for_plot_outcome_sampling_MCCFR.csv")
df2 = pd.read_csv("output/database_for_plot_NFSP.csv")
df3 = pd.read_csv("output/database_for_plot_FSP.csv")


df = pd.concat([df1, df2, df3], axis=1, join='outer')

df = df.loc[:,~df.columns.duplicated()]
df = df.set_index('iteration')

df
```

```python
plt.rcParams["font.size"] = 40
plt.figure(figsize=(24,12))

plt.plot(df.index, df.exploitability_FSP, label="FSP" , lw=5, color="dodgerblue")
plt.plot(df.index, df.exploitability_NFSP, label="NFSP", lw=5, color="magenta")
plt.plot(df.index, df.exploitability_outcome_sampling_MCCFR	, label="MCCFR" , lw=5, color="limegreen")

plt.xscale('log')
plt.yscale('log')
plt.xlabel('iteration')
plt.ylabel('exploitability')
plt.legend(loc='lower left')
plt.grid()

plt.xlim(10)
plt.ylim(0.1,10)
plt.savefig("./png/leduc_poker_nfsp.png")
```
